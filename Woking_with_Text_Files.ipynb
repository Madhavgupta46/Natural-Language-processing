```python
import pandas as pd
books={'Author':['Twain','Feyman','Hamilton'],'Topic':['Rafting','Physics','Mythology'],'Pages':[601,95,144]}
b=pd.DataFrame(books)
for i in b:
  print(f'{b}')
  break
```

```python
library=[('Author','Topic','Pages'),('Twain','Rafting',601),('Feyman','Physics',95),('Hamilton','Mythology',144)]
for book in library:
  print(f'{book[0]:{10}} {book[1]:{10}} {book[2]:.>{6}}')
```

## Date Format

```python
from datetime import datetime

today=datetime(year=2025, month=9,day=27)
print(f'{today:%B %d ,%Y}')
```

```python
today=datetime(year=2025, month=9,day=27)
print(f'{today:%b %d ,%Y}')
```
